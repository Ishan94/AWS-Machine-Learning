## Exploratory data analysis and feature creation
### What did the exploratory analysis find and how did you add additional features?
In the Exploratory Data Analysis, we find the following

* Datetime attribute of the dataset has several useful information like day, month, hour, etc. which can be used in making predictions, hence we extract these details and store it as additional attributes

* Certain attributes like season, weather, holiday, working day are categorical in nature and were earlier being treated as integers, hence we changed these to Categorical Attributes

### How much better did your model preform after adding additional features and why do you think that is?
There was a significant improvement in performance of the Model post the exploratory analysis. The Kaggle score changed from 1.39406 to 0.45956

This is because post the exploratory analysis, we have releavnt predictors like hour, day, time, etc. to make train and make predictions. Also transforming some attributes to categorical variables ensures their appropriate treatment thereby improving the model accuracy
